1.3 Body Mass Index
    Metric System (kg and m)
    BMI = weight in kilograms/(height in meters)^2
Enter the BMI Index:34
Very Overweight

Underweight: <18.5
Healthy / Normal Weight: 18.5 - 24.9
Overweight: 25.0 - 29.9
Obesity Class 1: 30.0 - 34.9
Obesity Class 2: 35.0 - 39.9
Extreme Obesity : >= 40

```python
#BMI calculation based on weight (in kg) and height ( in cm)
height=int(input("Please enter your height in centi meter"))
#print((height/100)** 2)
weight=int(input("please enter your weight in kg"))
bmi=weight/((height/100)** 2)
print("Your BMI is ", bmi)

#classification based on BMI

if(bmi <18.5):
    print("You are Underweight")
elif(bmi < 24.9):
    print("You are Healthy . Your are in Normal Weight. Keep continue your current diet")
elif(bmi < 29.9):
    print("You are Overweight. Need to take action on your diet")
elif(bmi < 34.9):
    print("You are Obesity Class 1. Need to do regular exercise along with diet ")
elif(bmi < 39.9):
    print("You are Obesity Class 2. Need to consult your doctor")
else:
    print("You are Extreme Obesity. Need a immediate doctor consultation ")


        

   
           
```
