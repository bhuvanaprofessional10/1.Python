4.Write a Python function to find the factorial of 5. Expected Output:
120



```python
#find factorial using math module
import math

fact_num=int(input("Enter the number for which factorial needs to be calculated"))
result=math.factorial(fact_num)
print(f"Factorial value of {fact_num} is {result}") 
```

```python
def factorial_value(factorial_number):
    result=1
    for f in range(1,factorial_number +1 ):
        result *= f

    print(result)

factorial_number=int(input("enter factorial number"))
factorial_value(factorial_number)
```

```python
#factorial using recursive 

def rec_factorial(n):
    if n < 0:
        return 0
    elif n ==0 or n ==1:
        return 1
    else:
        return n * rec_factorial(n-1)

number=int(input("enter the number to check the factorial using recursive"))
rec_factorial(number)
```

```python
#using math module: find a prime number , looping technique

import math
def isprime():
    num=int(input("enter your number input"))
    if(num <= 1):
        print("Given number is not eligible for prime number calculation. Numbers should be greater than 1 for calculation.")
        return False
    elif(num ==2):
        print(f"{num} is a prime number")
        return True
    else:
        print("range end by ",int(math.sqrt(num))+1)
        for i in range(2,int(math.sqrt(num))+1):
           # print("check =", i)
            if (num % i == 0):
                return False
                break
        
    return True
    
```
