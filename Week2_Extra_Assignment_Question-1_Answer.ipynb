```python
#add,subtract,multiply, divide
number1=int(input("enter the first number"))
number2=int(input("enter the second number"))

#add
add= number1+number2
print("             ")
print("number1=",number1)
print("number2=",number2)
print("add = ", add)

#subtract 
sub= number1-number2
print("             ")
print("number1=",number1)
print("number2=",number2)
print("subract = ", sub)

#multiply
mul= number1 * number2
print("             ")
print("number1=",number1)
print("number2=",number2)
print("multiply = ", mul)

#division
div= number1 / number2
print("             ")
print("number1=",number1)
print("number2=",number2)
print("Division = ", div)


#floor division
floorDiv= number1 // number2
print("             ")
print("number1=",number1)
print("number2=",number2)
print("Floor Division = ", floorDiv)

#modulo
mod= number1 % number2
print("             ")
print("number1=",number1)
print("number2=",number2)
print("modulo = ", mod)

#power

power= number1 ** number2
print("             ")
print("number1=",number1)
print("number2=",number2)
print("power = ", power)
```
