```python
"""
6.	Scenario: A unique lottery system assigns ticket numbers where only Armstrong numbers win the jackpot.
 Write logic to check whether a given number is an Armstrong number.

"""
number=int(input("Enter the number to check Armstrong"))
power_value=len(str(number))
print(power_value)

summed_value = sum(int(n) ** power_value for n in str(number))
   
if(number ==summed_value):
    print("given number is a Armstrong Number")
else:
    print("given number is NOT a Armstrong Number")
```
