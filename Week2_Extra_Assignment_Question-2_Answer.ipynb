```python
# print Odd Numbers in the list
number_list = [20,10,16,19,25,1,276,188]
for num in number_list:
    if(num % 2 >0 ):
        print(num," is odd")
```

```python
# print even Numbers in the list
number_list = [20,10,16,19,25,1,276,188]
for num in number_list:
    if(num % 2 ==0 ):
        print(num," is even")
```
